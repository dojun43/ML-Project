```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#한글폰트 설정
plt.rc('font', family = 'Malgun Gothic')
```

```python
train = pd.read_csv('train.csv', encoding='cp949')
test = pd.read_csv('test.csv', encoding='cp949')
```

```python
col_name = ['건물번호', '날짜', '전력사용량', '기온', '풍속','습도' ,'강우량', '일조', '비전기냉방설비운영', '태양광보유']
train.columns = col_name
train
```

## 전력 사용량 

```python
plt.figure(figsize=(20,5))

x_values = train.loc[train['건물번호'] == 1, '전력사용량'].index
y_values = train.loc[train['건물번호'] == 1, '전력사용량']
plt.plot(x_values,y_values)

x_values = train.loc[train['건물번호'] == 1, '전력사용량'].index
y_values = train.loc[train['건물번호'] == 20, '전력사용량']
plt.plot(x_values,y_values)

x_values = train.loc[train['건물번호'] == 1, '전력사용량'].index
y_values = train.loc[train['건물번호'] == 40, '전력사용량']
plt.plot(x_values,y_values)

x_values = train.loc[train['건물번호'] == 1, '전력사용량'].index
y_values = train.loc[train['건물번호'] == 60, '전력사용량']
plt.plot(x_values,y_values)

plt.legend(['건물1', '건물20', '건물40', '건물60'])
plt.title('전력사용량')
plt.show()
```

```python
train["월"] = train["날짜"].str.slice(5,7).astype(int)
df1 = train.groupby("월")["전력사용량"].mean()

names = ["6월","7월","8월"]
weights = df1.values
plt.barh(names,weights)
plt.title('월별 전력사용량 평균')
plt.show()
```

## 비전기냉방설비운영

```python
df_num = train.loc[train.날짜 == '2020-06-01 00']
```

```python
label = ['운영', '미운영']
values = list(df_num['비전기냉방설비운영'].value_counts().values)
explode = [0.05, 0.05]

plt.pie(values, labels=label, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True)
plt.title('비전기냉방설비운영')
plt.show()
```

## 태양광보유

```python
label = ['미보유', '보유']
values = list(df_num['태양광보유'].value_counts().values)
explode = [0.05, 0.05]

plt.pie(values, labels=label, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True)
plt.title('태양광보유')
plt.show()
```
